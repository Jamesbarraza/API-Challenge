```python
# Dependencies and Setup
#!pip install citipy
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from datetime import datetime
import seaborn as sns
from scipy.stats import linregress
import json
# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    #city = city.replace(" ", "%20")
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "Imperial"
# city = ""
city_name = []
cloudiness = []
country = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
wind_speed = []

record = 1

query_url = f"{url}appid={weather_api_key}&units={units}&q="
params = {
    "appid": weather_api_key,    
    "q": "cities",
    "units": units
}
print(f'Beginning Data Retrieval \n..............................')
for city in cities:
    params["q"] = city
    response = (requests.get(url, params)).json()
    try:
        city_name.append(response["name"])
        cloudiness.append(response["clouds"]["all"])
        country.append(response["sys"]["country"])
        date.append(response["dt"])
        humidity.append(response["main"]["humidity"])
        lat.append(response["coord"]["lat"])
        lng.append(response["coord"]["lon"])
        max_temp.append(response["main"]["temp_max"])
        wind_speed.append(response["wind"]["speed"])
        city_record = response["name"]
        print(f"Processing Record {record} of {len(cities)} complete. | Added {city_record}")
        print(response.url)
        record = record + 1
    except:
        print("City not found. Skipping...")
        record = record + 1
    continue
    

```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
output_data = {
    "City": city_name,
    "Cloudiness": cloudiness,
    "Country": country,
    "Date": date,
    "Humidity": humidity,
    "Lat": lat,
    "Lng": lng,
    "Max Temp": max_temp,
    "Wind Speed": wind_speed
}

cities_df = pd.DataFrame(output_data)

cities_df.to_csv("../output_data/my_cities.csv", encoding="utf-8")

cities_df.head() 


new_df = pd.DataFrame(
    {"City": city_name,
    "Cloudiness": cloudiness,
    "Country": country,
    "Date": date,
    "Humidity": humidity,
    "Lat": lat,
    "Lng": lng,
    "Max Temp": max_temp,
    "Wind Speed": wind_speed}
             )
new_df["Cloudiness"] = new_df["Cloudiness"].map("{:.2f}%".format)
new_df["Humidity"] = new_df["Humidity"].map("{:,.2f}%".format)
new_df["Max Temp"] = new_df["Max Temp"].map("{:.2f} (F)".format)
new_df["Wind Speed"] = new_df["Wind Speed"].map("{:,.2f} (mph)".format)
new_df
```

```python
# Create Northern and Southern Hemisphere DataFrames
north_df = cities_df.loc[cities_df["Lat"] >= 0.0]
south_df = cities_df.loc[cities_df["Lat"] < 0.0]

south_df.count()
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
(slope, intercept, rvalue, pvalue, stderr) = linregress(north_df["Lat"],north_df["Max Temp"])
regress_values = north_df["Lat"] * slope + intercept
print(f"The r-squared is: {rvalue}")
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
lblpos = (-5,5)
plt.scatter(north_df["Lat"],north_df["Max Temp"], marker="o", s=10)
plt.plot(north_df["Lat"],regress_values,"r-")
plt.title(f"(NH) Max Temperature vs. Latitude ({datestr})")
plt.annotate(line_eq,lblpos,fontsize=15,color="red")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.ylim(-20, 110)
plt.xlim(-10, 90)
plt.grid(True)

plt.show()
```
